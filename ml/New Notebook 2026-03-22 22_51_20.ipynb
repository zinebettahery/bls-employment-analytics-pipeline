```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# READ FROM SILVER TABLE
df_spark = spark.read.table("bls_cew.silver.cleaned_data")

# SELECT COLUMNS
df_spark_small = df_spark.select(
    "year",
    "annual_avg_emplvl"
)

# CONVERT TO PANDAS
df = df_spark_small.toPandas()

print(df.head())
# prepare data
df_ml = df[['year','annual_avg_emplvl']].dropna()

# group by year
df_ml = df_ml.groupby('year')['annual_avg_emplvl'].sum().reset_index()

# X / y
X = df_ml[['year']]
y = df_ml['annual_avg_emplvl']

# model
model = LinearRegression()
model.fit(X, y)

# predict next years
future_years = pd.DataFrame({'year':[2025,2026,2027]})
predictions = model.predict(future_years)

future_years['predicted_employment'] = predictions
print(future_years)
```
